## Analysis A: Term 1 2022 SPLC Members vs. Capture IT as of 28.02.2020

Import the files into dataframes:

```python
import pandas as pd
dfcm = pd.read_csv('splc-T1-2022.csv')
dfci = pd.read_csv('captureit.csv')
```

Confirming dataframes shows 537 Term 1 2022 SPLC members and 288 Capture IT learners:

```python
dfcm
```

```python
dfci
```

Make the field names consistent, "First Name" vs. "Given", "Last Name" vs. "Surname":

```python
dfci = dfci.rename(columns={'First Name':'Given','Last Name':'Surname'})
```

Confirm field names are now consistent:

```python
dfci.head()
```

```python
dfcm.head()
```

Find the overlap between Term 2 2022 members and Capture IT learners:

```python
dfovrlp = pd.merge(dfci,dfcm,how='inner',on=['Given','Surname'])
```

```python
dfovrlp
```

Result: **94 learners found**. I.e. out of 537 Term 1 2022 members, 94 are learners on Capture IT.
