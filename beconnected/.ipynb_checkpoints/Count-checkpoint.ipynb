```python
import pandas as pd
import openpyxl
from pathlib import Path
input_file = Path.cwd()/'beconnected.xlsx'
df = pd.read_excel(input_file,engine='openpyxl')
df.head()
```

```python
dfg = df.groupby(['Surname','First Name']).count()
```

```python
dfg.head()
```

```python
dfg.to_csv('hurrah.csv')
```

```python
input_file2 = Path.cwd()/'beconnected2020.csv'
df2020 = pd.read_csv(input_file2)
```

```python
df2020
```

```python
dfovrlp = pd.merge(dfg,df2020,how='inner',on=['First Name','Surname'])
```
