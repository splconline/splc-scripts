# Be Connected Progress Update

`beconnected.xlsx`: Latest spreadsheet of learner sessions

`captureit202X`: Capture IT data as of 202X

Read `beconnected.xlsx`, the list of sessions for the latest period:

```python
import pandas as pd
import openpyxl
from pathlib import Path
input_file1 = Path.cwd()/'beconnected.xlsx'
df = pd.read_excel(input_file1,engine='openpyxl')
df.head()
```

Group the sessions by person into the dataframe `dfg`:

```python
dfg = df.groupby(['Surname','First Name']).count()
```

```python
dfg.head()
```

This is the list of people who attended Be Connected for the latest period.

```python
dfg.to_csv('hurrah.csv')
```

Now import the list of everyone who is currently on Capture IT:

```python
input_file2 = Path.cwd()/'captureit2021.csv'
df2021 = pd.read_csv(input_file2)
```

```python
df2021
```

Rename the column from `Last Name` to `Surname` so it's easier to do the merge

```python
df2021 = df2021.rename(columns={'Last Name':'Surname'})
```

```python
df2021
```

Find out who are in the current sessions, who are already on Capture IT:

```python
dfovrlp = pd.merge(dfg,df2021.rename(columns={'Last Name':'Surname'}),how='inner',on=['First Name','Surname'])
```

```python
dfovrlp
```

Calculate the difference, this is the number of people who attended the current session, who are not on Capture IT:

```python
len(dfg.index) - len(dfovrlp.index)
```

```python
dfovrlp.to_csv('overlap.csv')
```
