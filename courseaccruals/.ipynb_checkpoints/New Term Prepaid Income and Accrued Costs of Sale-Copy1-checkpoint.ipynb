```python
import pandas as pd
from pathlib import Path
```

```python
input_file = Path.cwd()/'course_income.csv'
```

```python
df = pd.read_csv(input_file)
```

```python
df
```

Lose unwanted columns then sort by Course Code

```python
df.drop(['Day', 'Start',
       'Tot-Stus', 'Concession-Stus', 'CM Fee',
       'Concession Fee', 'Materials','First Session','Sessions','Operating Fee'], axis=1, inplace=True)
df = df.sort_values('Code')
```

```python
df
```

Make the '$1,200' strings numeric

```python
df['Category']=""
df=df[['Category','Code','Course','Total Income']]
```

```python
df
```

Populate the category field

```python
df['Category'] = df['Code'].replace('(\d+)','',regex=True)
```

```python
df
```

```python
df=df.sort_values(['Category','Course'])
```

```python
df.info()
```

```python
df.to_csv('hurrah.csv',index=False)
```
